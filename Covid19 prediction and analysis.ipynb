### Polynomial Regression


```python
from sklearn.preprocessing import PolynomialFeatures
X_poly=PolynomialFeatures(degree=2)
X_poly_train=X_poly.fit_transform(X_train)
X_poly_test=X_poly.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
poly=LinearRegression()
poly.fit(X_poly_train,Y_train)
```

```python
p=poly.score(X_poly_test,Y_test)*100
list_of_score.append(p)
```

```python
Yp=poly.predict(X_poly_test)
```

```python
plt.plot(Y_test,label='original',color='r')
plt.plot(Yp,label='predicted',color='k')
plt.rc('figure',figsize=(9,3))
plt.legend()
plt.show()
```

### SVR

```python
from sklearn.preprocessing import StandardScaler
Sc_X=StandardScaler()
Sc_X_train=Sc_X.fit_transform(X_train)
Sc_X_test=Sc_X.transform(X_test)
```

```python
Sc_Y=StandardScaler()
Sc_Y_train=Sc_Y.fit_transform(Y_train.reshape(-1,1))
SC_Y_test=Sc_Y.transform(Y_test.reshape(-1,1))
```

```python
from sklearn.svm import SVR
sv=SVR(kernel='rbf')
sv.fit(Sc_X_train,Sc_Y_train.ravel())
```

```python
s=sv.score(Sc_X_test,SC_Y_test.ravel())*100
list_of_score.append(s)
s
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=0)
dt.fit(X_train,Y_train)
```

```python
d=dt.score(X_test,Y_test)*100
list_of_score.append(d)
d
```

```python
Yp=dt.predict(X_test)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=300,random_state=0)
rfr.fit(X_train,Y_train)
```
